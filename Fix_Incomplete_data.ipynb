```python
import pandas as pd
import numpy as np
# Create dummy data
stuff = {
    "A": [1,2,3],
    "B": [4,np.nan,6],
    "C": [7,8,9],
    "D": [10,11,12]
}

df = pd.DataFrame(stuff)
df
```

```python
# Drop rows with null data
df.dropna()
```

```python
# Drop column with null data
df.dropna(axis=1)
```

```python
# Not permanent, inplace = False
# More than one? Set threshold
df.dropna(thresh=2, axis =1)
```

```python
# Replace things with fillna()
df.fillna(value="I'm filled")
```

```python
# Use Math functions
df.fillna(value=df["B"].mean())
```

```python
# Use min max
df.fillna(value=df["B"].min())
```
