```python
import pandas as pd
import numpy as np
# Create dummy data
# Practice dictionary for grouping
data = {
    "Corporation": [
        "Apple", "Google", "Microsoft", "Meta",
        "Tesla", "Amazon", "Netflix", "Walmart",
        "Pfizer", "Moderna"
    ],
    "CEO": [
        "Tim Cook", "Sundar Pichai", "Satya Nadella", "Mark Zuckerberg",
        "Elon Musk", "Andy Jassy", "Ted Sarandos", "Doug McMillon",
        "Albert Bourla", "Stéphane Bancel"
    ],
    "Industry": [
        "Technology", "Technology", "Technology", "Technology",
        "Automotive", "Retail", "Entertainment", "Retail",
        "Pharmaceutical", "Pharmaceutical"
    ],
    "Region": [
        "US", "US", "US", "US",
        "US", "US", "US", "US",
        "US", "US"
    ],
    "Founded": [
        1976, 1998, 1975, 2004,
        2003, 1994, 1997, 1962,
        1849, 2010
    ],
    "Annual_Revenue_Billion": [
        394.3, 280.9, 232.2, 116.6,
        81.5, 524.9, 31.6, 611.3,
        100.3, 18.5
    ]
}
# Create dataframe
my_df = pd.DataFrame(data)
my_df
```

```python
# Group by corporation
industry = my_df.groupby("Industry")["Annual_Revenue_Billion"]
```

```python
industry.sum()
```

```python
# mean
industry.mean()
```

```python
# max
industry.max()
```
