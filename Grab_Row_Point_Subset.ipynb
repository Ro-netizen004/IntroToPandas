```python
import pandas as pd
import numpy as np
my_df = pd.read_csv("dog_data.csv")
my_df
```

```python
# Grab Row Method 1 Location
my_df.loc[0]
```

```python
# Grab Row Method 2 Location
my_df.iloc[0]
```

```python
# Specific Points (Rows and Columns)
# my_df.loc["Row", "Column"]
my_df.loc[0, "Breed"]
```

```python
# Subsets
# my_df.loc[['row', 'row'], ['column', 'column']]
my_df.loc[[1,3],["DogName","Breed"]]
```

Conditional Selection starts from this point onwards

```python
# Conditional Selection
my_df == "BROWN"
```

```python
# Returns Dataframe with data
my_df[my_df == "BROWN"]
```

```python
# Return them on a column
my_df[my_df == "BROWN"]["Color"]
```

```python
my_df[my_df == "BROWN"][["Color", "Breed"]]
```

Multiple conditionals start from here onwards.

```python
# Multiple conditionals AND
my_df[(my_df["Color"] == "BROWN") & (my_df["Breed"] == "MIXED")]
```

```python
# Return a specific column
my_df[(my_df["Color"] == "BROWN") & (my_df["Breed"] == "MIXED")]["DogName"]
```
