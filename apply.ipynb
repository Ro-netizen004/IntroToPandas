```python
import pandas as pd
import numpy as np

data = {
    "Corporation": [
        "Apple", "Google", "Microsoft", "Meta",
        "Tesla", "Amazon", "Netflix", "Walmart",
        "Pfizer", "Moderna"
    ],
    "CEO": [
        "Tim Cook", "Sundar Pichai", "Satya Nadella", "Mark Zuckerberg",
        "Elon Musk", "Andy Jassy", "Ted Sarandos", "Doug McMillon",
        "Albert Bourla", "Stéphane Bancel"
    ],
    "Industry": [
        "Technology", "Technology", "Technology", "Technology",
        "Automotive", "Retail", "Entertainment", "Retail",
        "Pharmaceutical", "Pharmaceutical"
    ],
    "Region": [
        "US", "US", "US", "US",
        "US", "US", "US", "US",
        "US", "US"
    ],
    "Founded": [
        1976, 1998, 1975, 2004,
        2003, 1994, 1997, 1962,
        1849, 2010
    ],
    "Annual_Revenue_Billion": [
        394.3, 280.9, 232.2, 116.6,
        81.5, 524.9, 31.6, 611.3,
        100.3, 18.5
    ]
}

my_df = pd.DataFrame(data)
my_df
```

```python
# Create function
def timesbillion(x):
  return x*1000000000

# Use apply function
pd.DataFrame(my_df["Annual_Revenue_Billion"].apply(timesbillion))
```

```python
# Use lambda
pd.DataFrame(my_df["Annual_Revenue_Billion"].apply(lambda x: x*1000000000))
```

```python
# Sort Revenue from lowest to highest
my_df.sort_values(by="Annual_Revenue_Billion", inplace=True)
my_df
```

```python
my_df.sort_values(by="Annual_Revenue_Billion", inplace=True, ascending=False)
my_df
```
